First worked on print_menu method by using for loop with enumerate function to traverse the tuple containing the details of menu items, price and quantity. To display the menu starting from index 1, added 1 to the key.

Next worked on is_valid_login method by blocking the irrelevant code from the main() to check the output results for any errors. For this method, used while loop with condition count less than 3 by initiating count equal to zero at first and used input function to get userid and password details from the user. Next the program used if condition to check the userid exists in self.managers dictionary containing the managers userid and password details, if it exists then the program checks for the password matches the userid in self.managers, then print 'success' and sets that manager as self.current_manager in case of successful login. If the entered userid or password doesn't matches with key-value pair in self.managers, program prints 'Access denied Try again' and increases the count by 1 for every unsuccessful login. After three unsuccessful logins, program prints 'system closed due to incorrect tries!'

Next worked on take_order method using update function on the self.items_ordered dictionary by taking item and quantity ordered as key-value pair into the empty dictionary. Checked for the output results for any errors by unblocking the relevant code in the main().

Next worked on process_order by assigning 0 to subtotal and using for loop in self.items_ordered.items() to get the subtotal value of all items ordered by multiplying the item price with quantity for each item ordered.

Next worked on print_sales_receipt method to display the order summary by using for loop to traverse the list of items ordered along with its price, quantity and item total. Next displays the sub total value using process_order function, tax amount by assigning 10% of subtotal value to tax and finally total amount by adding subtotal and tax values.

```python
class Restaurant:

    def __init__(self, name, menu):
        self.name = name
        self.menu = menu
        self.items_ordered = {}
        # managers: id and password pairs
        # i.e., id: lion and password: king
        self.managers = {"james":"bond", "lion":"king", "peter":"pan"}
        self.current_manager = ""

    def is_valid_login(self):
        """ is valid login:
            take input from user, id and pw
            check id exists in self.managers
            check pw matches key value in self.managers
            max 3 tries
            return True or False
            if Ture, set slef.current_manager if successful login
        """
        count = 0    
        while count < 3:
            userid = input("User ID:")
            userpw = input("Password:")
            
            if userid in self.managers:      #checks for userid in the dictionary
                if userpw == self.managers[userid]:    #checks for the password match with the id in the dictionary
                    print("Success!")
                    self.current_manager = userid      #assigns the manager as current manager in case of true
                    return True  
                else:
                    print("Access denied. Try again...")
                    count += 1                         #increases the count for every unsuccessful login
            else:
                print("Access denied. Try again...")
                count += 1
        return False
            
            
    def print_menu(self):
        """ print menu after successful login:
        
            --SUISP Ordering System--
            Wlecome to, SHAKE SHOCK BURGER!
            Current manager, Lion.

            ------------
            The menu
            ------------
            [1] Shock Burger $4.85
            [2] French Fries $2.85
            [3] Milk Shake $4.5
        """
        print()
        print("--SUISP Ordering System-- \n Wlecome to, {}! \n Current manager, {}.".format(self.name, \
                                                                                            self.current_manager))
        print()
        print("----------")
        print(" The Menu ")
        print("----------")
        # traverses the menu items with its price
        for key, val in enumerate(self.menu):             
            print ("[{}] {} ${}".format(key+1, val[0], val[1]))
            
            
    def take_order(self, item, quantity):
        """ take order:
            item number and quantity
            update attribute self.items_ordered, i.e., {1:2, 3:1}
        """
        self.items_ordered.update({item : quantity}) #updates the dictionary with item number and quantity ordered
        
        
    def process_order(self):
        """ process order:
            use attribute, items_ordered
            calculate sub total price on all items
            return sub total
        """
        subtotal = 0
        for item, quantity in self.items_ordered.items():
            subtotal += self.menu[item][1]*quantity   #calculates the subtotal by multiplying item price with quantity
 
        return subtotal
            
        
    def print_sales_receipt(self):
        """ print sales receipt on this order:
            ++++++++++++++++++++++++
            The current order summary:
            ++++++++++++++++++++++++
            Shock Burger $4.85
                @ 3: item total $14.55
            French Fries $2.85
                @ 1: item total $2.85
            Milk Shake $4.50
                @ 5: item total $22.50
            ---------------------
            Sub Total: $39.90
            Tax: $3.99
            Total: $43.89
        """
        print()
        print("*************************")
        print("The current order summary")
        print("*************************")
        
        # displays the list of items ordered with its price, quantity and total value
        for item, quantity in self.items_ordered.items():
            print ("{} ${:.2f} \n    @ {}: item total ${:.2f}".format(self.menu[item][0], \
                                                    self.menu[item][1], quantity, self.menu[item][1]*quantity))
        print("-------------------------")
        subtotal = self.process_order()
        tax = subtotal * 0.10            
        print("Sub Total: ${:.2f}".format(subtotal))
        print("Tax: ${:.2f}".format(tax))
        print("Total: ${:.2f}".format(subtotal + tax))
        
        
            
def main():
    menu = ["Shock Burger", float(4.85), 50], ["French Fries", float(2.85), 200], ["Milk Shake", float(4.50), 20]
    rt = Restaurant('Shake Shock Burger', menu)

    if rt.is_valid_login():
        rt.print_menu()
        more_items = "yes"
        while more_items == "yes":
            choice, quantity = int(input("\nWhat would you like?\n")), int(input("How many would you like?\n"))
            if 0 <= (choice-1) < len(rt.menu):
                rt.take_order(choice-1, quantity)
            more_items = (input("Do you want to order more items?")).lower()
        rt.print_sales_receipt()
    else:
        print("System closed due to incorrect tries!")
        

if __name__ == '__main__':
    main()
```
